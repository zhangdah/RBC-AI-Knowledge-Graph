# 1.0 Load Data

```python
# import required module
import os
import re
# assign directory
directory = '/content/drive/MyDrive/soilmoisture_csv'
def get_file (directory): 
# iterate over files in
# that directory
  filelist = []
  for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # print(filename)
    # checking if it is a file
    if os.path.isfile(f):
        print(f)
        soilmoisture = pd.read_csv(f)
        soilmoisture['field_id'] = re.findall(r'\d+', filename)[0]
        filelist.append(soilmoisture)
  
    #print(filelist) 

  return pd.concat(filelist)
```

```python
df = get_file (directory)
# df = df[pd.to_datetime(df['date'])<= pd.to_datetime('2015-03-01')]
df = df[pd.to_datetime(df['date'])  <= '2015-06-01']

df.reset_index(inplace = True)
```

# 3.0 Convert each row of dataframe to triples

```python
for index, row in df.iloc[:,:].iterrows():
    #print(row['Date/Time'])
    #print(index)
    # g.add((URIRef(flamel+'StationBoundary'+str(index+1)), RDF.type, flamel.Boundary))
    # g.add((URIRef(flamel+'Station'+str(index+1)), flamel.hasLatitude, Literal(row['Latitude'], datatype=XSD.float)))
    # g.add((URIRef(flamel+'Station'+str(index+1)), flamel.hasLongitude, Literal(row['Longitude'], datatype=XSD.float)))
    # g.add((URIRef(flamel+'Station'+str(index+1)), flamel.hasCoordinate, Literal(row['Coordinates'], datatype=geo.wktLiteral)))
    # g.add((URIRef(flamel+'Station'+str(index+1)), flamel.hasName, Literal(row['Name'], datatype=XSD.string)))

    field_id = row['field_id']
    date = str(row['date']).replace("-", "_")
    subject = "_"+"field"+ field_id +"_"+date

    # SoilMostureContent_fieldi_xxxx -typeOf-> SoilMoistureContent class
    g.add((URIRef(flamel+'SoilMoistureContent'+subject), RDF.type, flamel.HistoricalSoilMoistureContent))

    # Precipitation MostureContent_fieldi_xxxx -affect-> SoilMostureContent_fieldi_xxxx
    #g.add((URIRef(flamel+'Precipitation'+subject), flamel_2.affect, URIRef(flamel+'SoilMoistureContent'+subject)))
    g.add((URIRef(flamel+'Precipitation'+subject), flamel.affect, URIRef(flamel+'SoilMoistureContent'+subject)))

    # SoilMostureContent_fieldi_xxxx -hasAmount-> SoilMoistureContent value
    #g.add(( URIRef(flamel+'SoilMoistureContent'+subject), flamel_2.hasAmount, Literal(row['average'], datatype=XSD.float)))
    g.add(( URIRef(flamel+'SoilMoistureContent'+subject), flamel.hasAmount, Literal(row['average'], datatype=XSD.float)))

    # SoilMostureContent_fieldi_xxxx -hasDateTime-> SoilMoistureContent value
    g.add(( URIRef(flamel+'SoilMoistureContent'+subject), time.inXSDDate, Literal(pd.to_datetime(row['date']).date(), datatype=XSD.date)))
```
